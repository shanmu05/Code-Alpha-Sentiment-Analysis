```python

import pandas as pd

file_name = 'Reviews.csv'

try:
    df = pd.read_csv(file_name, engine='python', on_bad_lines='warn')

    print("âœ… File loaded successfully!")
    display(df.head())

except Exception as e:
    print(f"âŒ An error occurred: {e}")
    print("\nTroubleshooting Tips:")
    print("1. Please double-check that you downloaded the correct CSV file and not the webpage.")
    print(f"2. Make sure the filename '{file_name}' in the code exactly matches the uploaded file.")
print("Data Info:")
df.info()
reviews_df = df[['Text', 'Score']].copy()

reviews_df.dropna(inplace=True)

reviews_df.rename(columns={'Text': 'review_text', 'Score': 'rating'}, inplace=True)

print("Cleaned Data:")
display(reviews_df.head())

print("\nCleaned Data Info:")
reviews_df.info()
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd 

nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

reviews_df['sentiment_scores'] = reviews_df['review_text'].apply(lambda text: sia.polarity_scores(str(text)))

print("Data with Sentiment Scores:")
display(reviews_df.head())
reviews_df['compound_score'] = reviews_df['sentiment_scores'].apply(lambda score_dict: score_dict['compound'])

print("\nData with Final Compound Score:")
display(reviews_df.head())
def classify_sentiment(score):
    if score > 0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

reviews_df['sentiment_type'] = reviews_df['compound_score'].apply(classify_sentiment)

print("Data with Sentiment Classification:")
display(reviews_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

plt.figure(figsize=(7, 5))
sns.countplot(x='sentiment_type',
              data=reviews_df,
              order=['Positive', 'Neutral', 'Negative'],
              palette='viridis')

plt.title('Overall Sentiment of Amazon Reviews', fontsize=16)
plt.xlabel('Sentiment Type', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.show()
```
